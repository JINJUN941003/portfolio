```python
from geopy.geocoders import Nominatim

def geocoding(address):
    geolocoder = Nominatim(user_agent = 'USA', timeout=None)
    geo = geolocoder.geocode(address)
    crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}

    return crd
```

```python
city = list(NY["city"])
city_set = set(city)
```

```python
cities = dict()
for i in city_set:
    try:
        cities[i] = geocoding(i)
    except:
        continue 
```

```python
cities
```

```python
NY_cities = cities
```

```python
df = pd.DataFrame(data= NY_cities)
```

```python
NY_cities = np.transpose(df)
NY_cities
```

```python
NY_cities = NY_cities.reset_index()
NY_cities.columns = ['city', 'lat','lng'] 
```

```python
NY_Join = pd.merge(NY,NY_cities, left_on='city', right_on='city', how='inner')
NY_Join
```

```python
NY_Join["lat"] = NY_Join["lat"].astype("float")
NY_Join["lng"] = NY_Join["lng"].astype("float")
```

```python
NY_real_final = NY_Join[(NY_Join["lat"] >= 40.494444) & (NY_Join["lat"] <= 45.011667) & (NY_Join["lng"] <= -71.790278 )& (NY_Join["lng"] >= -79.765000)]
```
